# drop unnamed column

```python
df = df.drop("Unnamed: 0",axis =1)
df
```

# gender distribution

```python
plt.figure(figsize=(5,5))
ay=sns.countplot(data = df , x ="Gender")
ay.bar_label(ay.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
gp = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gp
```

```python
sns.heatmap(gp, color = 'Green',annot =True)
plt.show()
```

# from the above chart we concluded that the parent education score affect their student score

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gp1
```

```python
sns.heatmap(gp1, color = "green",annot =True)
plt.show()
```

# from the above chart we conclude that there is no such impact on the student's score

```python
print(df["EthnicGroup"].unique())
```

```python
groupA= df.loc[(df["EthnicGroup"] == "group A")].count()
groupB= df.loc[(df["EthnicGroup"] == "group B")].count()
groupC= df.loc[(df["EthnicGroup"] == "group C")].count()
groupD= df.loc[(df["EthnicGroup"] == "group D")].count()
groupE= df.loc[(df["EthnicGroup"] == "group E")].count()


l=["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct = "1.2f%%")
plt.show()
```
